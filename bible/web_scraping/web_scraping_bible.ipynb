```python
# Función para añadir un nuevo capítulo
def anadir_capitulo(ruta_archivo, nuevo_capitulo):
    # Leer el archivo JSON existente
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        datos = json.load(archivo)
    
    # Añadir el nuevo capítulo a la lista de capítulos
    datos['capitulos'].append(nuevo_capitulo)
    
    # Guardar los datos actualizados de nuevo en el archivo JSON
    with open(ruta_archivo, 'w', encoding='utf-8') as archivo:
        json.dump(datos, archivo, ensure_ascii=False, indent=4)
```
